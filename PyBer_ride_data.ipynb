```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_df=pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_df
```

```python
pyber_df.plot(x='Month', y='Avg. Fare ($USD)')
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_df))
tick_locations=[value for value in x_axis]
# Plot the data.
pyber_df.plot(x='Month', y='Avg. Fare ($USD)')
plt.xticks(tick_locations, pyber_df['Month'])
plt.show()
```

```python
pyber_df.plot.bar(x='Month', y='Avg. Fare ($USD)')
plt.show()
```

```python
import statistics
y= pyber_df['Avg. Fare ($USD)']
stdev= statistics.stdev(y)
stdev
```

```python
pyber_df.plot(x='Month', y='Avg. Fare ($USD)',yerr=stdev,capsize=3, kind='bar', color='skyblue')
plt.xticks(rotation=0)
plt.yticks(np.arange(0, max(y)+5, 5.0))
plt.show()
```
